```python

with open("support_tickets.txt", "w") as f:
    f.write(data)
```

```python
import fasttext
```

```python
model = fasttext.train_supervised(input="support_tickets.txt",lr=0.5,epoch=25,wordNgrams = 2,dim =100,loss="softmax")
```

```python
model.save_model("model.ftz")
```

```python
labels, probs = model.predict("otp not received")
print(labels, probs)
```
